```python
from collections import deque

# ─────────────────────────────────────────────
# Struktur Data: Doubly Linked List Node
# ─────────────────────────────────────────────
class Node:
    def __init__(self, id_kotak, kalori):
        self.id_kotak = id_kotak
        self.kalori   = kalori
        self.prev     = None
        self.next     = None


class DoublyLinkedList:
    def __init__(self):
        self.head = None
        self.tail = None

    # Insert di paling kanan (Insert Last)
    def insert_last(self, id_kotak, kalori):
        node = Node(id_kotak, kalori)
        if self.tail is None:
            self.head = self.tail = node
        else:
            node.prev      = self.tail
            self.tail.next = node
            self.tail      = node

    # Representasi maju (head → tail)
    def forward_str(self):
        parts, curr = [], self.head
        while curr:
            parts.append(f"{curr.id_kotak}({curr.kalori})")
            curr = curr.next
        return " <-> ".join(parts)

    # Representasi mundur (tail → head)
    def backward_str(self):
        parts, curr = [], self.tail
        while curr:
            parts.append(f"{curr.id_kotak}({curr.kalori})")
            curr = curr.prev
        return " <-> ".join(parts)

    # Bubble Sort ascending berdasarkan kalori (tukar data, bukan node)
    def sort_ascending(self):
        if not self.head:
            return
        swapped = True
        while swapped:
            swapped = False
            curr = self.head
            while curr and curr.next:
                if curr.kalori > curr.next.kalori:
                    curr.id_kotak,       curr.next.id_kotak = \
                        curr.next.id_kotak, curr.id_kotak
                    curr.kalori,         curr.next.kalori   = \
                        curr.next.kalori,   curr.kalori
                    swapped = True
                curr = curr.next

    # Kembalikan list dari tail → head (untuk pemuatan bagasi)
    def tail_to_head(self):
        result, curr = [], self.tail
        while curr:
            result.append((curr.id_kotak, curr.kalori))
            curr = curr.prev
        return result


# ─────────────────────────────────────────────
# Baca Input
# ─────────────────────────────────────────────
n = int(input())
orders = []
for _ in range(n):
    parts = input().split()
    orders.append((parts[0], int(parts[3])))   # (ID_Kotak, Kalori)

q = int(input())
instructions = [input().strip() for _ in range(q)]


# ══════════════════════════════════════════════
# FASE 1: Antrean Dapur (Queue / FIFO)
# ══════════════════════════════════════════════
print("=== FASE 1: Antrean Dapur ===")

queue = deque()
for order in orders:
    print(f"{order[0]} masuk antrean.")
    queue.append(order)

print("Semua pesanan diproses ke meja inspeksi.")
print()

# ══════════════════════════════════════════════
# FASE 2: Meja Inspeksi (Doubly Linked List)
# ══════════════════════════════════════════════
print("=== FASE 2: Meja Inspeksi ===")

dll = DoublyLinkedList()
while queue:
    item = queue.popleft()
    dll.insert_last(item[0], item[1])

print(f"Maju: {dll.forward_str()}")
print(f"Mundur: {dll.backward_str()}")
print()

# ══════════════════════════════════════════════
# FASE 3: Pengurutan Kalori (Bubble Sort ↑)
# ══════════════════════════════════════════════
print("=== FASE 3: Pengurutan Kalori ===")

dll.sort_ascending()
print(f"Maju: {dll.forward_str()}")
print()

# ══════════════════════════════════════════════
# FASE 4: Pemuatan Mobil Boks (Stack/Deque LIFO)
# Kotak kalori terbesar (ujung kanan) dimuat pertama → ke dasar bagasi
# Deque: kiri = terluar, kanan = terdalam
# ══════════════════════════════════════════════
print("=== FASE 4: Pemuatan Mobil Boks ===")

bagasi = deque()
for item in dll.tail_to_head():          # mulai dari kalori terbesar
    bagasi.appendleft(item)

bagasi_str = ", ".join(f"{x[0]}({x[1]})" for x in bagasi)
print(f"Isi Bagasi (Terluar ke Terdalam): {bagasi_str}")
print()

# ══════════════════════════════════════════════
# FASE 5: Pembagian di Sekolah (Double-ended Queue)
# NORMAL  → ambil dari terluar (kiri deque / atas bagasi)
# KHUSUS  → ambil dari terdalam (kanan deque / dasar bagasi)
# ══════════════════════════════════════════════
print("=== FASE 5: Pembagian di Sekolah ===")

for instr in instructions:
    if instr == "NORMAL":
        item = bagasi.popleft()
        print(f"Bagikan NORMAL: {item[0]}({item[1]}) keluar.")
    else:   # KHUSUS
        item = bagasi.pop()
        print(f"Bagikan KHUSUS: {item[0]}({item[1]}) keluar.")

sisa_str = ", ".join(f"{x[0]}({x[1]})" for x in bagasi)
print(f"Sisa di Bagasi: {sisa_str}")
```
